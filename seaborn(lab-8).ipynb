
### **Matplotlib**:
1. **Foundation for Data Visualization**: Matplotlib is a widely used 2D plotting library in Python, providing a foundation for creating a variety of charts like line plots, bar charts, scatter plots, and more.
2. **Highly Customizable**: It offers extensive control over plot elements, including colors, axes, labels, and more, making it highly customizable but requiring more manual work for complex plots.
3. **Object-Oriented and Pyplot Interface**: It offers two main interfaces â€“ an object-oriented API for more control and the simpler `pyplot` interface, which functions like MATLAB.
4. **Integration with Other Libraries**: Matplotlib integrates well with other Python libraries, such as Pandas, making it easy to plot data directly from datafsks in Python.

### **Seaborn**:
1. **Statistical Data Visualization**: Seaborn is built on top of Matplotlib and focuses on making statistical plots more intuitive and informative with built-in themes and color palettes.
2. **High-Level Abstraction**: It provides a high-level interface for creating more complex visualizations like heatmaps, violin plots, and pair plots with less code compared to Matplotlib.
3. **Integration with Pandas**: Seaborn works seamlessly with Pandas dataframes, making it easy to plot categorical and multi-dimensional data.
4. **Default Aesthetics**: Seaborn comes with attractive default styles and color schemes, producing visually appealing plots without needing much customization.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

## General Statistics Plot

```python
sns.pairplot(data=iris,hue='species',palette="muted")
plt.show()
```

The plot provides a grid of scatterplots and density plots to explore how these features vary between three species of iris flowers (Setosa, Versicolor, and Virginica), represented by different colors:

Blue: Setosa
Orange: Versicolor
Green: Virginica
Summary:
Diagonal Plots: The diagonal of the grid shows the distribution (Kernel Density Estimation, KDE) of individual features for each species. For instance:

Setosa flowers have distinct, narrower distributions for petal length and petal width compared to the other species.
Versicolor and Virginica overlap in petal and sepal dimensions, but Virginica tends to have larger values.
Scatter Plots: The off-diagonal plots compare pairs of features, where each dot represents a flower:

Setosa (blue) is well-separated from the other species in terms of petal length and petal width.
Versicolor (orange) and Virginica (green) show some overlap, particularly in petal width, but differ in their petal length distribution.
This pairplot helps in identifying how well different features can distinguish between the species and shows that petal dimensions are more effective in separating Setosa from the other two species, while Versicolor and Virginica are harder to distinguish.

## Pie Plot for Species Frequency:

```python
plt.pie(iris["species"].value_counts(),labels=iris["species"].value_counts().index,autopct="%1.2f%%")
plt.show()
```

## Relation between sepal length and sepal width

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,hue="species",palette="muted")
plt.show()
```

## Distribution of Sepal and Petal Features

```python
sns.pairplot(data=iris,hue='species',palette="muted")
plt.show()
```

## Jointplot of Sepal Length vs Sepal Width

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=iris,hue="species",palette="muted")
plt.show()
```

## KDE Plot for Setosa Species (Sepal Length vs Sepal Width)

```python
setosa=iris[iris["species"]=="setosa"]
```

```python
sns.kdeplot(x="sepal_length",y="sepal_width",data=setosa,shade=True,hue="species",palette="pastel")
plt.title("KDE Plot for Setosa Species (Sepal Length vs Sepal Width)")
plt.show()
```

## KDE Plot for Setosa Species (Petal Length vs Petal Width)

```python
sns.kdeplot(x="petal_length",y="petal_width",data=setosa,shade=True,hue="species",palette="pastel")
plt.title("KDE Plot for Setosa Species (Petal Length vs Petal Width)")
plt.show()
```

While Matplotlib is ideal for detailed, customizable plots, Seaborn simplifies the process of creating visually appealing statistical plots.
